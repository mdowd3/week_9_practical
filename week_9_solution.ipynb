```r
#Question 8
mlm_model = lm(richness~ angle1 + angle2 + exposure + salinity + temperature + NAP + penetrability + grainsize + humus + chalk + sorting1 , data = survey_data_richness)
summary(mlm_model)
```

Based on model above, only NAP is relevant predictor, as it is the only one with a significant p-value

```r
#Question 9
stepAIC(mlm_model)
```

Based on the stepAIC function, the model with the best fit compares richness to angle2, salinity, temperature, NAP, grain size, and chalk. Using this method over the previous method (using all variables) is better becuase some of the data interacts with one another in such a way to confound the regression. stepAIC tests different permutations of data to find the one with the best fit that doesnt result in confoudning data.
